```python

data = pd.read_csv("./House_prediction (5).csv",header=0, na_values='-')
```

```python
columns = list(data.columns)
```

```python
for col in columns[1:8]:
  means_of_groups = data.groupby([col]).describe()[[(columns[8], 'mean'),(columns[10], 'mean'),(columns[11], 'mean')]]

  fig, ax = plt.subplots(1, 3, figsize=[9.6, 4.8])
  for i,j in zip([8,10,11], range(3)):
    ax[j].plot(np.array(means_of_groups.index), means_of_groups[(columns[i], 'mean')])
    ax[j].set(title=f"{columns[i]} V/S {col}")
  fig.show()
```

Now these graphs are pretty inconclusive. 
So we can remove anomalies in data (extraordinarily high values) and plot graphs city-wise and put limits in order to analyze only the highly concentrated data-zones.

```python
columns = ["city", "area","rooms","bathroom","parking spaces","floor","animal","furniture","hoa (R$)","rent amount (R$)","property tax (R$)","fire insurance (R$)","total (R$)"]
df = pd.read_csv("House_prediction (5).csv",usecols=columns)
df.columns = df.columns.str.replace('[#,@, ,(R$) ]', '')
print("Contents in csv file:\n", df)
```

```python
df_filter_pt = df.loc[(df['city'] == 'Rio de Janeiro') & (df['propertytax'] < 1000) ]

plt.plot(df_filter_pt.propertytax, df_filter_pt.rooms)
plt.xlabel('propertytax')
plt.ylabel('rooms')
plt.show()
plt.plot(df_filter_pt.propertytax, df_filter_pt.bathroom)
plt.xlabel('propertytax')
plt.ylabel('bathroom')
plt.show()
plt.plot(df_filter_pt.propertytax, df_filter_pt.area)
plt.xlabel('propertytax')
plt.ylabel('area')
plt.show()
plt.plot(df_filter_pt.propertytax, df_filter_pt.parkingspaces)
plt.xlabel('propertytax')
plt.ylabel('parkingspaces')
plt.show()
plt.plot(df_filter_pt.propertytax, df_filter_pt.floor)
plt.xlabel('propertytax')
plt.ylabel('floor')
plt.show()
plt.plot(df_filter_pt.propertytax, df_filter_pt.furniture, 'ro')
plt.xlabel('propertytax')
plt.ylabel('furniture')
plt.show()
plt.plot(df_filter_pt.propertytax, df_filter_pt.animal, 'ro')
plt.xlabel('propertytax')
plt.ylabel('animal')
plt.show()
```

```python
df_filter_hoa = df.loc[(df['city'] == 'Rio de Janeiro') & (df['hoa'] < 3500) ]

plt.plot(df_filter_hoa.hoa, df_filter_hoa.rooms)
plt.xlabel('hoa')
plt.ylabel('room')
plt.show()
plt.plot(df_filter_hoa.hoa, df_filter_hoa.bathroom)
plt.xlabel('hoa')
plt.ylabel('bathroom')
plt.show()
plt.plot(df_filter_hoa.hoa, df_filter_hoa.area)
plt.xlabel('hoa')
plt.ylabel('area')
plt.show()
plt.plot(df_filter_hoa.hoa, df_filter_hoa.parkingspaces)
plt.xlabel('hoa')
plt.ylabel('parkingspaces')
plt.show()
plt.plot(df_filter_hoa.hoa, df_filter_hoa.floor)
plt.xlabel('hoa')
plt.ylabel('floor')
plt.show()
plt.plot(df_filter_hoa.hoa, df_filter_hoa.furniture, 'ro')
plt.xlabel('hoa')
plt.ylabel('furniture')
plt.show()
plt.plot(df_filter_hoa.hoa, df_filter_hoa.animal, 'ro')
plt.xlabel('hoa')
plt.ylabel('animal')
plt.show()
```
