```python
import numpy as np
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['House_prediction.csv']),header=0, na_values='-')
```

```python
display(df)
```

```python
df.info()
df['floor']=pd.to_numeric(df['floor'])
df.info()
```

```python
df2=df.groupby('city')
df3=df2.agg(['mean'])
display(df3)
```

```python
for i in range(1,10):
  df4 = df3.iloc[:,i-1]
  plt.plot(df4)
  plt.title(df4.name[0])
  plt.grid()
  plt.show()
```
